```python
import pandas as pd
```

```python
df_people = pd.read_csv('C:/Users/human/미정프로젝트/서울시 상권분석서비스(길단위인구-상권).csv',
                        encoding='cp949')

df_area = pd.read_csv('C:/Users/human/미정프로젝트/서울시 상권분석서비스(영역-상권).csv',
                      encoding='cp949')

df_sales = pd.read_csv('C:/Users/human/미정프로젝트/서울시 상권분석서비스(추정매출-상권).csv',
                       encoding='cp949')

print(f'길단위인구: {len(df_people):,}행')
print(f'영역: {len(df_area):,}행')
print(f'추정매출: {len(df_sales):,}행')
```

```python
print('=== 길단위인구 컬럼 ===')
print(df_people.columns.tolist())
print()
print('=== 영역 컬럼 ===')
print(df_area.columns.tolist())
print()
print('=== 추정매출 컬럼 ===')
print(df_sales.columns.tolist())
```

```python
# 영역에서 필요한 컬럼만 추출
df_area_slim = df_area[['상권_코드', '상권_코드_명', '자치구_코드_명', '행정동_코드_명', '영역_면적']]

# 길단위인구 + 영역 합치기
df_merged = pd.merge(df_people, df_area_slim, on='상권_코드', how='left')

print(f'합치기 전: {len(df_people):,}행')
print(f'합치기 후: {len(df_merged):,}행')
print()
print(df_merged[['상권_코드', '자치구_코드_명', '총_유동인구_수', '연령대_20_유동인구_수']].head())
```

```python
# 추정매출 상권별 평균으로 집계
df_sales_agg = df_sales.groupby('상권_코드').agg(
    평균_당월매출=('당월_매출_금액', 'mean'),
    평균_주중매출=('주중_매출_금액', 'mean'),
    평균_주말매출=('주말_매출_금액', 'mean'),
    연령대_20_매출=('연령대_20_매출_금액', 'mean'),
    연령대_30_매출=('연령대_30_매출_금액', 'mean'),
    연령대_40_매출=('연령대_40_매출_금액', 'mean'),
    연령대_50_매출=('연령대_50_매출_금액', 'mean'),
).reset_index()

# 합치기
df_merged = pd.merge(df_merged, df_sales_agg, on='상권_코드', how='left')

print(f'합치기 후: {len(df_merged):,}행')
print(f'컬럼 수: {len(df_merged.columns)}개')
df_merged.head(3)
```

```python
# 핵심 컬럼만 선택
df_final = df_merged[[
    '자치구_코드_명',
    '총_유동인구_수',
    '연령대_20_유동인구_수',
    '연령대_30_유동인구_수',
    '연령대_40_유동인구_수',
    '연령대_50_유동인구_수',
    '연령대_60_이상_유동인구_수',
    '시간대_11_14_유동인구_수',
    '시간대_17_21_유동인구_수',
    '평균_당월매출',
    '평균_주중매출',
    '평균_주말매출',
    '연령대_20_매출',
    '연령대_30_매출',
    '연령대_40_매출',
    '연령대_50_매출',
    '영역_면적'
]].copy()

# 구별 평균으로 집계
df_gu = df_final.groupby('자치구_코드_명').mean().reset_index()
df_gu = df_gu.rename(columns={'자치구_코드_명': '구명'})

print(f'행 수: {len(df_gu)}개 (25개 구)')
df_gu.head()
```

```python
# 1번에서 저장한 폐업률 파일 불러오기
df_closure = pd.read_csv('C:/Users/human/미정프로젝트/폐업률_구별_최근5년.csv', encoding='utf-8-sig')

# 구명 기준으로 합치기
df_ml = pd.merge(df_gu, df_closure[['구명', '폐업률']], on='구명', how='left')

print(f'결측치 확인:')
print(df_ml['폐업률'].isnull().sum())
print()
print(df_ml[['구명', '총_유동인구_수', '평균_당월매출', '폐업률']].head(10))
```

```python
df_closure = pd.read_csv('C:/Users/human/미정프로젝트/새 폴더/폐업률_구별_연도별_업종별.csv', encoding='utf-8-sig')

# 구 + 연도 + 업종 기준으로 합치기
df_ml_final = pd.merge(df_sales_gu, df_closure[['구명', '연도', '업종', '폐업률']],
                       on=['구명', '연도', '업종'], how='left')

print(f'행 수: {len(df_ml_final):,}개')
print(f'결측치: {df_ml_final["폐업률"].isnull().sum()}개')
print(df_ml_final.head())
```

```python
# 결측치 있는 행 확인
print('결측치 있는 조합:')
print(df_ml_final[df_ml_final['폐업률'].isnull()][['구명', '연도', '업종']].head(10))

# 결측치 제거
df_ml_final = df_ml_final.dropna(subset=['폐업률'])

print(f'\n결측치 제거 후: {len(df_ml_final):,}개')
```

```python
# 결측치 구 평균으로 채우기
df_ml_final['폐업률'] = df_ml_final.groupby(['구명', '업종'])['폐업률'].transform(
    lambda x: x.fillna(x.mean())
)

print(f'결측치 남은 개수: {df_ml_final["폐업률"].isnull().sum()}')
```

```python
df_subway2 = pd.read_csv('C:/Users/human/미정프로젝트/서울_상권분석_용_정제_결측치처리_최종_데이터.csv',
                         encoding = 'utf-8-sig')
print(f'행 수: {len(df_subway2):,}개')
print(df_subway2[['사용월', '지하철역', '자치구']].head())
```

```python
#연도 추출
df_subway2['연도'] = df_subway2['사용월'].astype(str).str[:4].astype(int)
df_subway2 = df_subway2[(df_subway2['연도'] >= 2019) & (df_subway2['연도'] <= 2024)]

print(f'필터링 후 행 수: {len(df_subway2):,}개')
print(f'연도 범위: {sorted(df_subway2["연도"].unique())}')
```

```python
# 시간대별 컬럼 합산해서 총 승차/하차 계산
승차컬럼 = [col for col in df_subway2.columns if '승차인원' in col]
하차컬럼 = [col for col in df_subway2.columns if '하차인원' in col]

df_subway2['총_승차'] = df_subway2[승차컬럼].sum(axis=1)
df_subway2['총_하차'] = df_subway2[하차컬럼].sum(axis=1)

# 구별 + 연도별 평균 집계
df_subway_gu = df_subway2.groupby(['자치구', '연도']).agg(
    평균_승차=('총_승차', 'mean'),
    평균_하차=('총_하차', 'mean'),
).reset_index()

df_subway_gu = df_subway_gu.rename(columns={'자치구': '구명'})

print(f'행 수: {len(df_subway_gu):,}개')
print(df_subway_gu.head())
```

```python
# 최종 ML 데이터 불러오기
df_ml_final = pd.read_csv('최종_ML용_데이터.csv', encoding='utf-8-sig')

# 구 + 연도 기준으로 합치기
df_ml_final = pd.merge(df_ml_final, df_subway_gu, on=['구명', '연도'], how='left')

print(f'행 수: {len(df_ml_final):,}개')
print(f'결측치: {df_ml_final.isnull().sum().sum()}개')
print(df_ml_final.columns.tolist())
```

```python
df_ml_final.to_csv('최종_ML용_데이터.csv', index=False, encoding='utf-8-sig')
print('저장 완료!')
```
